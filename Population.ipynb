```python
years = df.columns[4:]  
total_population = df.loc[df['Country Name'] == 'World', years].values.flatten()
plt.figure(figsize=(15, 8))
years = list(map(int, years))
plt.bar(years, total_population, color = 'green')
plt.xticks([year for year in range(1960,2021,5)])
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('World Population (1960-2022)')

plt.show()
```

```python
years = df.columns[4:]  
total_population = df.loc[df['Country Name'] == 'World', years].values.flatten()
plt.figure(figsize=(15, 8))
years = list(map(int, years))
plt.barh(years, total_population, color = 'skyblue')
plt.xticks([year for year in range(1960,2021,5)])
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('World Population (1960-2022)')

plt.show()
```

```python
years = df.columns[4:]  
total_population = df.loc[df['Country Name'] == 'World', years].values.flatten()
plt.figure(figsize=(15, 8))
years = list(map(int, years))
plt.plot(years, total_population, color = 'r', marker = 'o')
plt.xticks([year for year in range(1960,2021,5)])
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('World Population (1960-2022)')

plt.show()
```
